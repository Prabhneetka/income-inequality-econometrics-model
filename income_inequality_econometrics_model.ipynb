# Determinants of Income Inequality

This project investigates the economic factors influencing income inequality using econometric modeling techniques.

The analysis demonstrates the use of regression models, Maximum Likelihood Estimation, and Generalized Method of Moments.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

```

```python
gini = [30,32,35,40,42,38,36,34,33,31]
education = [5,6,7,4,3,6,7,8,7,6]
gdp = [2000,2500,3000,1500,1200,2800,3200,3500,3300,2900]
unemployment = [6,7,8,10,11,7,6,5,6,7]

data = pd.DataFrame({
    "Gini":gini,
    "Education":education,
    "GDP":gdp,
    "Unemployment":unemployment
})

data
```

```python
plt.scatter(data["Education"],data["Gini"])
plt.xlabel("Education Level")
plt.ylabel("Income Inequality (Gini)")
plt.title("Education vs Inequality")
plt.show()
```

```python
X = data[["Education","GDP","Unemployment"]]
Y = data["Gini"]

X = sm.add_constant(X)

model = sm.OLS(Y,X).fit()

print(model.summary())
```

## Econometric Interpretation

The regression model estimates the relationship between education levels, economic development, unemployment, and income inequality.

Higher education levels are expected to reduce inequality, while unemployment may increase income disparities.
